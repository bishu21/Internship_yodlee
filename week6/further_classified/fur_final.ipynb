```python
import pandas as pd
```

```python
broken=pd.read_excel("further.xlsx",sheet_name='data')
```

```python
broken=broken[['name','existence']]
```

```python
broken
```

```python
def gas(x):
    l=x['name']
    if("gas" in l)|("Gas" in l)|("Oil" in l):
        return "YES"
    else:
        return "NO"
```

```python
broken['Gas']=broken.apply(gas,axis=1)
```

```python
broken
```

```python
def elect(x):
    l=x['name']
    if(" Elect" in l)|("Electric" in l )|("electric" in l):
        return "YES"
    else:
        return "NO"
```

```python
broken['Electricity']=broken.apply(elect,axis=1)
```

```python
broken
```

```python
def water(x):
    l=x['name']
    if("WATER" in l) | ("water" in l)|("Water" in l)|("Sewer" in l)|("Irrigation" in l):
        return "YES"
    else:
        return "NO"
```

```python
broken['water']=broken.apply(water,axis=1)
```

```python
broken
```

```python
broken.to_csv("final.csv")
```
